## after_dataのimagesフォルダーに含まれる不要なfile.jpgを削除
* 01_transform_data.ipynbでデータの整形を行った際になぜかimageフォルダの最後に不要な.jpgが生成されている  
* 例えば、trainのimg406.jpgは空のファイルなのでそれらを削除

```python
# 削除対象の.jpgがあるフォルダまでのpath
target_dir = "./after_data/train/images"

# 削除対象のfile
filename = "img406.jpg"

# 対象フォルダ内のすべてのファイルを処理
for root, subdir, files in os.walk(target_dir):
    print(root, subdir, filename)
    for file in files:
        print(file)
        # ファイル名がfilenameと一致する場合は削除
        if file == filename:
            os.remove(os.path.join(root, file))
            print(f"空のファイルを削除しました。：{os.path.join(root, file)}")
```
